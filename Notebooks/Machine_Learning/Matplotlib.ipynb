## Scatter Plot

***Used to find OUTLIERS***

```python
plt.scatter(df['Data.Temperature.Avg Temp'],df['Data.Precipitation'],c=['black'])

plt.xlabel('Avg Temp')
plt.ylabel('Precipitation')
```

```python
plt.scatter(df['Date.Month'],df['Data.Precipitation'],c=['black'])

plt.xlabel('Date.Month')
plt.ylabel('Data.Precipitation')
```
