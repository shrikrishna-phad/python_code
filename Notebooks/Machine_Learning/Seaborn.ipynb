# sns.jointplot( x = ' ', y = ' ' , data = ' ' , kind = ' hex / reg ' )

***combination of scatterplot and histagram***

**kind = 'reg' ---> gives us standerd daviation line on histagram and mean line on the scatterplot**

**hue = 'categorical_col_name'**
    
    sometimes kind and hue doesnot work simultenously
    
    This can be used for detection outliers in the data
    ---> in given example we can safely assume that 

```python
sns.color_palette()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg',color="#34495e")
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='scatter',hue='day')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hist',color='gold')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='resid',color='silver')
```

# sns.pairplot ( data = data , hue = ' ', kind = ' ' )


**To plot multiple pairwise bivariate distributions in the given dataset**

can get relation between all numerical columns in dataset

```python
sns.pairplot(df.groupby('day').get_group('Sun'),hue='sex')

plt.show()
```

```python
sns.pairplot(df.groupby('day').get_group('Sat'),hue='sex')

plt.show()
```

```python
sns.pairplot(df.groupby('day').get_group('Thur'),hue='sex')

plt.show()
```

# sns.heatmap ( )

***Gives the colored representation of co-relation between numerical columns***


    if we provide values for vmin & vmax then it will show only data between those values

```python
sns.heatmap(df.corr(),annot=True,vmax=1,vmin=0)
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```
