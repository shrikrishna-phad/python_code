# ***Read***  'r'

1. ---> If file is preasent it will read the content.
2. ---> If file is not preasent then it will show an exception- Filenotfound error

3. ---> If mode is not specified in fp = open('file_name','mode') then it will take read ('r') as the default mode.

**Methods of Read**

1. .read(n)      ---> To read 'n' charecters from file.
                 ---> n is always a positive integer, if we gave negative value it will show whole file.
2. .readline()   ---> To read any one line at a time.
                 ---> To read m^th line we need to give that command m times.
                 ---> If we gave n in readlines(n) then it will behave same as .read(n)
                 ---> Type(fp.readline()) --> 'str'
3. .readlines()  ---> To read all lines in list.
                 ---> It will make a list of all lines considering earch '\n' as end of line.
                 ---> It will create a list of strings, each sentance as a element of list.

```python
fp = open('resume.txt') #Default Read mode
data = fp.read(6) # first 6 charecters 01234, considers space ' ' as a charecter
fp.close()
print(data)
```

```python
fp = open('resume.txt') #Default Read mode
data = fp.read(-1) # n is always a positive integer, if we gave negative value it will show whole file.
print(data)
fp.close()
```

# Append + Read   'a+'

***Append then Read***

1. --> File preasent --> Write content at the end.
                     --> No Overwrite.

2. --> File not Preasent --> Create file then write.

```python
fp = open('resume.txt','a+')
fp.write('Hi from append + mode')
d1 = fp.read()
print('d1= :', d1)   # Does not print because cursor is at the end of line

print(fp.tell()) # shows cursor position

print(fp.seek(0)) # moves cursor at specified position

d2 = fp.read()
print('d2= :', d2)
fp.close()
```

# Exclusive  'x+'

1. --> File Preasent --> FileExistsError

2. --> File Not preasent --> Creats file then writes.

***not readable***

    **WRITE mode**                                **EXCLUSIVE mode**
    
    --> File preasent                             --> File preasent
        OverWrite                                     FileExistsError
     
    --> File Not preasent                         --> File Not preasent
        Creates file then write                       Creates file then write

```python
fp = open('resume.txt','x')
fp.write('This is exclusive mode file')
fp.seek(0)
print(fp.read())
fp.close()
```

```python
fp = open('resumex.txt','x') # Created new file
fp.write('This is exclusive mode file') # Writes The given thing
fp.seek(0) # Moves cursor at 0th position
print(fp.read()) # Not supported read operation in this mode
fp.close()
```

***Q1. content in file 1 copy in file 2***

```python
with open('resume.txt','r') as fp1:
    with open('resume_copy.txt','w+') as fp2:
        d = fp1.read()
        print("fp1 :", d)
        fp2.write(d)
        fp2.seek(0)
        d1 = fp2.read()
        print("\nfp2 :", d1)
```

***Q2. file 1 content write in file 2 delete file 1***

```python
with open('resume_copy.txt','r+') as fp2:
    with open('resume_copy2.txt','w+') as fp3:
        d1 = fp2.read()
        fp3.write(d1)
        fp2.seek(0)
        fp2.truncate()
        fp2.seek(0)
        fp3.seek(0)
        
        d2 = fp2.read()
        d3 = fp3.read()
        
        print("1st file content before truncate : ", d1)
        print("\n1st file content after truncate : ", d2)
        print("\n2nd file content : ", d3)
```

**Q**

**'fun1' - to detect file preasent or not**

***if not present* --> go to 'fun2'**

**'fun2' - to create file then go to 'fun4'**

**'fun4' - to write new content**

***if present* --> go to 'fun3'**

**'fun3' - to overwrite**


```python
import os


def fun2(f):
    with open(f,'w') as fp:
        print('file created')
    print('in fun2')

def fun3(f):
    with open(f,'w') as fp:
        fp.write("Overwritten")
    print("in fun3")
    
def fun4(f):
    with open(f,'w') as fp:
        fp.write("NewContent")
    print("in fun4")
        
def fun1():
    file = input("Enter the file name (With Extension) :")
    if os.path.isfile(file):
        print(f"File {file} exists")
        fun3(file)
    else:
        print(f"File {file} does not exists")
        fun2(file)
        fun4(file)
        
fun1()
```

***Q. find number of lines, words and charecters in file***

```python
with open('resume.txt','r') as f:
    print('\nNumber of lines : ',len(f.readlines()))
    f.seek(0)
    d = f.read()
    print('\nNumber of words : ',len(d.split(" ")))
    f.seek(0)
    x = 0
    y = 0
    for i in d:
        x+=1
        if i not in {' '}:
            y+=1

print('\nCharecters including spaces : ',x)
print('\nCharecters excluding spaces : ',y)
```

# Handling csv file

***comma seprated values***

---> CSV module is not present in python automatically, so we need to import it.

***MODULE***---> Module is nothing but a python file.

---> Suppose we have a file named

         test.py                          test1.py
     a = 10                              import test
     b = 20                              add(10,20)
     def add(a,b):                        --> 30
         print(a+b)
         
---> Need to create object to perform operations on csv file, like obj = csv.writer/csv.reader 

```python
import csv

with open('test.csv','w') as fp:
    w_obj = csv.writer(fp)
    w_obj.writerow(['Name','Age','Address'])
    w_obj.writerow(['Keshav',25,'Pune'])

# Adds Extra line to each entry
```

```python
with open('test.csv','w',newline='') as fp:
    w_obj = csv.writer(fp)
    w_obj.writerow(['Name','Age','Address'])
    w_obj.writerow(['Keshav',25,'Pune'])
```

***Take input from user on how many rows he/she needs to add records***

```python
with open('test.csv','w',newline='') as fp:
    wobj = csv.writer(fp)
    wobj.writerow(['NAME','AGE','ADDRESS'])
    
    no_row = int(input("Enter the number of rows needed :"))
    
    for i in range(no_row):
        n = input("Enter the name :")
        a = input("Enter the age :")
        add = input("Enter the address :")
        wobj.writerow([n,a,add])
```

***Read CSV***

```python
with open('test.csv','r') as fp:
    robj = csv.reader(fp)
    data = list(robj) # List of lines in a list
    data1 = tuple(robj)
    print(data)
    print(data1)
```

```python
with open('test.csv','r') as fp:
    robj = csv.reader(fp)
    data1 = tuple(robj) # List of lines in a tuple
    data = list(robj) # List of lines in a list
    print(data)
    print(data1)
```

***Print the CSV line by line***

```python
with open('test.csv','r') as fp:
    robj = csv.reader(fp)
    data = list(robj)
    print(data)
    
    for lines in data:
        for i in lines:
            print(i,end=' ')
        print(" ")
```

***Copy from one CSV to another CSV***

```python
with open('test.csv','r') as fp:
    robj = csv.reader(fp)
    data = list(robj)
    
    with open('csvcopy.csv','w') as f:
        wobj = csv.writer(f)
        
        for i in data:
            wobj.writerow(i)
```
