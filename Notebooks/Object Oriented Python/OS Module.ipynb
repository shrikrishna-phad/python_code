# OS

---> OS module is collection of functions that allows devloper to interact with operating system.

---> Need to import inbuilt os module.

```python
import os
```

```python
os.getcwd()  # To Get path of working current directory.
```

```python
os.chdir('D:')  # To change working directory.

os.getcwd()
```

```python
os.chdir('E:')

os.getcwd()
```

```python
os.mkdir('os module test directory') # To Create directory | Only one folder created.
```

```python
os.rmdir('os module test directory') # To Delete directory | Only one folder deleted.
```

```python
os.makedirs('os directory structure/sub structure1/sub structure2') # To Create directory structure (folder inside folder).
```

```python
os.removedirs('os directory structure/sub structure1/sub structure2') # To Delete directory structure (folder inside folder).
```

```python
os.listdir('os directory structure') # To get list of directories inside given directory.
```

```python
for dirpath,dirname,filename in os.walk('D:'):
    print(dirpath)
    print(dirname)
    print(filename)
```

```python
print(os.system('date'))
print(os.system('time'))
print(os.system('tree'))
```
