# Inheritance

---> Increases **REUSABILITY**

---> Adds **FEATURES**

---> Increases **SCALABILITY**



---> If in both classes method name is same then the priority will be to the method whose class object is created.

---> If in both classes have *init* method then the priority will be to the *init* method whose class object is created.


### Single Inheritance



```python
class A:
    def m1(self):
        print("in M1")
        
class B(A):
    def m2(self):
        print("in M2")
```

```python
obj_single = B()
obj_single.m1()
obj_single.m2()
```

### Multilevel Inheritance



```python
class A:
    def m1(self):
        print("in M1")

class B(A):
    def m2(self):
        print("in M2")
        
class C(B):
    def m3(self):
        print("in M3")

class D(C):
    def m4(self):
        print("in M4")
```

```python
obj_multilevel = D()

obj_multilevel.m1()

obj_multilevel.m2()

obj_multilevel.m3()

obj_multilevel.m4()
```

### Multiple Inheritance



```python
class A:
    def m1(self):
        print("in M1")

class B:
    def m2(self):
        print("in M2")
        
class C(A,B):
    def m3(self):
        print("in M3")
```

```python
obj_multiple = C()

obj_multiple.m1()
obj_multiple.m2()
obj_multiple.m3()
```

### Cyclic Inheritance



```python
class A(D):
    def m1(self):
        print("in M1")

class B(A):
    def m2(self):
        print("in M2")
        
class C(B):
    def m3(self):
        print("in M3")

class D(C):
    def m4(self):
        print("in M4")
```

```python
obj.cyclic_a = A()
obj.cyclic_b = B()
obj.cyclic_c = C()
obj.cyclic_d = D()
```

```python
obj.cyclic_a.m1()
obj.cyclic_a.m2()
obj.cyclic_a.m3()
obj.cyclic_a.m4()
```
