# Confusion Matrix

***Confusion matrix is a matrix (table) that can be used to measure the performance of an machine learning alghorithm***

---> Each row of confusion matrix repreasents an instance of Actual class and each column repreasents an instance of Predicted class (This can differ but for now we are considering this arengment).




##### Accuracy

It is Probability that *Actual Positive* will test positive and *Actual Negative* will test negative.

                     TN + TP                       True Positive + True Negative
    Accuracy = ------------------  = ---------------------------------------------------------
                TN + TP + FN + FP     True(Positive + Negative) + False(Positive + Negative) or All Datapoints
                
                
##### Recall / True Positive Rate / Sensitivity

It is Probability that *Actual Positive* will test Positive.

What proportion of actual positives was identified correctly?
    
---> Mostly used where False Negative (Type II error) trumps False Positive. (Type II error is more serious)

                 TP                 True Positive
    Recall = ---------- = ----------------------------------
              TP + FN        True Positive + False Negative or Actual Positive
              
              
              
##### True Negative Rate / Specificity

It is Probability that *Actual Negative* will test Negative.

             TN                  True Negative
    TNR = ---------- = ---------------------------------
           TN + FP       True Negative + False Positive or Actual Negative


##### Precision

What proportion of positive identifications was actually correct?

                 TP                True Positive
    Recall = ---------- = --------------------------------
              TP + FP      True Positive + False Positive or Predicted Positive
